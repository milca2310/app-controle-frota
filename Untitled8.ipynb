```python
# Passo 2: Importar as bibliotecas necess√°rias
import os
import sqlite3
import streamlit as st
from datetime import date
from pyngrok import ngrok

# Apagar o banco de dados existente (se for seguro fazer isso)
if os.path.exists('frota.db'):
    os.remove('frota.db')
    print("Banco de dados 'frota.db' exclu√≠do com sucesso.")
else:
    print("Banco de dados 'frota.db' n√£o encontrado.")

# Conectando ao banco de dados SQLite
conn = sqlite3.connect('frota.db', check_same_thread=False)
c = conn.cursor()

# Criando tabelas
c.execute('''CREATE TABLE IF NOT EXISTS rota (
    data TEXT,
    km_inicio INTEGER,
    km_fim INTEGER,
    tipo_movel TEXT,
    notas INTEGER,
    valor_diario REAL,
    problemas TEXT
)''')

c.execute('''CREATE TABLE IF NOT EXISTS abastecimento (
    data TEXT,
    litros REAL,
    valor_abastecimento REAL,
    km_atual INTEGER,
    tipo_carga TEXT
)''')

c.execute('''CREATE TABLE IF NOT EXISTS manutencao (
    tipo_manutencao TEXT,
    data_manutencao TEXT,
    quilometragem INTEGER,
    proxima_manutencao INTEGER
)''')

# Layout do Streamlit
st.markdown("""
    <h1 style='text-align: center; color: #795548;'>üìã Controle Di√°rio da Frota</h1>
    <h4 style='text-align: center; color: #A1887F;'>Registre tudo do seu dia em um s√≥ lugar</h4>
""", unsafe_allow_html=True)

abas = st.tabs(["üöö Rota do Dia", "üõ¢ Abastecimento", "üîß Manuten√ß√£o", "üì¶ Entregas", "üìä Relat√≥rios"])

def calcular_custo_por_km():
    # Buscando dados de abastecimento e rota
    c.execute('''SELECT SUM(abastecimento.valor_abastecimento), SUM(rota.km_fim - rota.km_inicio)
                 FROM abastecimento
                 JOIN rota ON abastecimento.km_atual = rota.km_fim''')
    resultado = c.fetchone()

    if resultado[0] is not None and resultado[1] is not None:
        valor_abastecimento_total, km_total = resultado
        if km_total > 0:
            custo_km = valor_abastecimento_total / km_total
            st.write(f"Custo por KM: R$ {custo_km:.2f}")
            if custo_km > 2:  # Alerta se o custo por KM for maior que R$2
                st.warning("‚ö† O custo por KM est√° muito alto! Verifique seus gastos.")
        else:
            st.warning("‚ö† N√£o h√° dados suficientes para calcular o custo por KM.")
    else:
        st.warning("‚ö† N√£o h√° dados suficientes de abastecimento ou rota para calcular o¬†custo¬†por¬†KM.")

def verificar_manutencao():
    c.execute("SELECT tipo_manutencao, quilometragem, proxima_manutencao FROM manutencao")
    manutencao = c.fetchall()
    for m in manutencao:
        tipo, km_atual, proxima_manutencao = m
        try:
            km_atual = int(km_atual)
            proxima_manutencao = int(proxima_manutencao)
        except ValueError:
            continue
        if km_atual >= proxima_manutencao:
            st.warning(f"‚ö† A manuten√ß√£o de {tipo} est√° atrasada! Quilometragem atual: {km_atual}. Pr√≥xima manuten√ß√£o em: {proxima_manutencao}.")

# Aba Rota do Dia
with abas[0]:
    st.subheader("Rota do Dia")
    data = st.date_input("Data", value=date.today(), key="rota_data")
    km_inicio = st.number_input("KM Inicial", min_value=0)
    km_fim = st.number_input("KM Final", min_value=0)
    tipo_movel = st.text_input("Tipo de M√≥vel Transportado")
    notas = st.number_input("N√∫mero de Notas no Dia", min_value=0)
    valor_diario = st.number_input("Valor Di√°rio Recebido (R$)", min_value=0.0)
    problemas = st.text_area("Problemas do Dia (devolu√ß√µes, quebras...)")

    if st.button("Salvar Rota"):
        c.execute('''INSERT INTO rota VALUES (?, ?, ?, ?, ?, ?, ?)''',
                  (str(data), km_inicio, km_fim, tipo_movel, notas, valor_diario, problemas))
        conn.commit()
        st.success("Rota registrada com sucesso!")

# Aba Abastecimento
with abas[1]:
    st.subheader("Registro de Abastecimento")
    data_ab = st.date_input("Data do Abastecimento", value=date.today(), key="abastecimento_data")
    litros = st.number_input("Litros Abastecidos", min_value=0.0, key="litros")
    valor_abastecimento = st.number_input("Valor Pago (R$)", min_value=0.0, key="valor_abastecimento")
    km_atual = st.number_input("Quilometragem Atual", min_value=0, key="km_atual")
    tipo_carga = st.text_input("Tipo de Carga (opcional)", key="tipo_carga")

    if st.button("Salvar Abastecimento"):
        c.execute('''INSERT INTO abastecimento VALUES (?, ?, ?, ?, ?)''',
                  (str(data_ab), litros, valor_abastecimento, km_atual, tipo_carga))
        conn.commit()
        st.success("Abastecimento registrado com sucesso!")

# Aba Manuten√ß√£o
with abas[2]:
    st.subheader("Registro de Manuten√ß√£o")
    tipo_manutencao = st.text_input("Tipo de Manuten√ß√£o")
    data_manutencao = st.date_input("Data da Manuten√ß√£o", value=date.today(), key="manutencao_data")
    quilometragem = st.number_input("Quilometragem", min_value=0)
    proxima_manutencao = st.number_input("Pr√≥xima Manuten√ß√£o (KM)", min_value=0)

    if st.button("Salvar Manuten√ß√£o"):
        c.execute('''INSERT INTO manutencao VALUES (?, ?, ?, ?)''',
                  (tipo_manutencao, str(data_manutencao), quilometragem, proxima_manutencao))
        conn.commit()
        st.success("Manuten√ß√£o registrada com sucesso!")

# Alertas
verificar_manutencao()
calcular_custo_por_km()

# Passo 4: Rodar o ngrok para gerar o link p√∫blico
public_url = ngrok.connect(8501)
print(f"Link p√∫blico do Streamlit: {public_url}")

# Rodar o Streamlit
!streamlit¬†run¬†app.py¬†&
```
